Adding a distance column from berlin center taking advantage of the lattitudes and longitude columns using geopy

```python
from geopy.distance import great_circle

def distance_from_berlin(lat, lon):
    berlin_centre = (52.5027778, 13.404166666666667)
    record = (lat, lon)
    return great_circle(berlin_centre, record).km

#add distanse dataset
listings1['distance_km'] = listings1.apply(lambda x: distance_from_berlin(x.latitude, x.longitude), axis=1)

del listings1['latitude']
del listings1['longitude']
listings1.head(1)
```

The object columns which have true and false values can be changed to numeric 1 and zero
those columns are require_guest_phone_verification,host_is_superhost,host_has_profile_pic,
                    host_identity_verified,is_location_exact,requires_license,instant_bookable

```python
def truth_table(ele):
    if ele == 'T' or ele == 't':
        return 1.0
    elif ele == 'F' or ele == 'f': 
        return 0.0
    else:
        return np.nan

tf_columns = ["require_guest_phone_verification","host_is_superhost","host_has_profile_pic",
                    "host_identity_verified","is_location_exact","requires_license","instant_bookable"]
for i in tf_columns:
    list_sum3[i] = list_sum3[i].map(truth_table)
```
